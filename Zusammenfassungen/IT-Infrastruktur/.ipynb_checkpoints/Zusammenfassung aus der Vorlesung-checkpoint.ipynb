<h3 style="color:#32CD32"> Einführung und Definition</h3>

Ein Betriebssystem beschreibt die Programme eines digitalen
Rechensystems, die zusammen mit den Eigenschaften der Rechenanlage die Basis der möglichen Betriebsarten bilden und insbesondere  die Abwicklung von Programmen steuern und überwachen. (Definition nach DIN 44300)

Das Betriebssystem (engl.: operating system, OS) umfasst die Gesamtheit der Systemprogramme eines Rechners, die (zusammen mit den Eigenschaften der Rechnerhardware) die grundlegende Infrastruktur für die Ausführung von Anwendungssoftware bilden.
Das Betriebssystem bildet eine Abstraktion von Hardwareeigenschaften und ist für die Steuerung und Überwachung von Anwendungsprogrammen zuständig. Es besteht aus einem Betriebssystemkern (engl.: operating system kernel) und aus Dienstprogrammen (engl. utility program).

Zusammenfassend: Sammlung von grundlegenden, systemnahen Programmen (Software) und Bibliotheken, die für andere Programme (Anwenderprogramme) eine Infrastruktur schaffen.

<h3 style="color:#32CD32">Klassifizierung und Nutzerperspektiven auf Betriebssysteme</h3>

<u>Aufgaben des Betriebssystems</u>:

1. Verwaltung von Ressourcen: Haupspeicher, CPU, ect.
2. Programmierschnittstelle: API
3. Bedienschnittstelle für Nutzer: User Interface 

Spezielle Betriebssysteme haben sich für spezifische Rechnertypen entwickelt, wie nachfolgende Abbildung veranschaulicht:

<img src="spezielleBetriebssysteme.png" alt="Quelle: Vorlesung Einführung in die Programmierung, FH Dortmund, Marcel Hoffmann " title="" />

Neben den speziellen Rechnertypen kann man Betriebssysteme noch wie folgt klassifizieren:

- Ein- oder Mehrbenutzersysteme (single/multi user os)
- Ein- oder Mehrprogrammsysteme (single task/multitasking os)
- offen oder proprietär (propriertär bedeutet: ein Hersteller, kein offener Source Code)
- spezialisierte Betriebsart:
    - Stapelverarbeitungs-Betriebssysteme (batch processing OS)
    - Dialogbetrieb-Betriebssysteme (interactive/dialog processing)
    - Netzwerk-Betriebssysteme (network OS)
    - Echtzeitverarbeitung (realtime OS)
- Desktop-, Workstation-, Server- und Embedded-Betriebssysteme

<u>Perspektiven bei Nutzung von Betriebssystemen:</u>
- Anwenderperspektive(Grundlage der anderen Sichten):
    - Verwaltung von Dateien und Verzeichnissen im Dateisystem
    - Nutzung externer Geräte
    - Prozessverwaltung
    - Bedienung des Kommandozeilen-Interpreters("Shell")
    - Selbsthilfe und Problembehebung
- Administrator Perspektive:
    - Durchführung der Installation, des Boot-Vorgangs und Systemstarts
    - Benutzerverwaltung
    - Konfiguration von System- und Netzwerkeinstellungen
    - Optimierung und Sicherstellung der Systemsicherheit
- Programmierperspektive:
    - Unix-Anwendungsentwickler: Umgang mit Systemaufrufen
    - Kernel-Entwickler: Spezialisierte Programmierung auf der Betriebssystemkernel-Ebene
    
Die Perspektiven verdeutlichen die verschiedenen Rollen und Anforderungen an ein Betriebssystem, die von der einfachen Dateiverwaltung über die Systemwartung bis hin zur Entwicklung von Systemsoftware reicht.

<h3 style="color:#32CD32">Aufgaben von Betriebssystemen</h3>

<u>Allgemein:</u> Das Betriebssystem ist dafür verantwortlich die Ressourcen des Rechnersystems zu verwalten und den Programmen zuzustellen. Diese Ressourcenverwaltung ist zentral für die effiziente und effektive Nutzung eines Computersystems

Doch was sind die ressourcen eines Rechnersystems?: Prozessoren, Arbeitsspeicher, Festplattenspeicher, Bildschirm, Drucker, Dateien, Netzwerkverbindungen, Kommunikationskanäle, Semaphoren, ect.

<u>Beispiele für Ressourcenverwaltung:</u>
- Prozessoren/Rechenzeit: Der Scheduler des Betriebssystems verteilt die Rechenzeit unter allen laufenden Programmen. Dies ermöglicht ein ausgewogenes und gerechtes ;amaement der CPU-Ressourcen.
- Drucker: Wenn mehrere Programme gleichzeitig auf den Drucker zugreifen wollen, sorgt das Betriebssystem dafür, dass die Zugriffe nacheinander erfolgen. Dies verhindert Konflikte und stellt sicher, dass jeder Druckauftrag korrekt abgearbeitet wird.

Anhand dieser Punkte sieht man, dass das Betriebssystem essenzielle für die Organisation und Verwaltung der vielfältigen und komplexen Ressourcen eines computer-Systems ist.

<h3 style="color:#32CD32">Prozessverwaltung</h3>

<u>Definition Prozess:</u> Ein Prozess ist eine Instanz eines Programms, das vom Betriebssystem ausgeführt wird. 
<u>Definition Multitasking:</u> Mehrere Prozesse können vom Betriebssystem entweder scheinbar oder tatsächlich parallel ausgeführt werden. Dieses Konzept nennt sich Multitasking.

<u><b>Charakteristiken eines Prozesses:</b></u>
- <b>Eigener Adressraum:</b> Jeder Prozess hat einen eigenen Adressraum, der von den Adressräumen anderer Prozesse isoliert ist
- <b>Eigene Ressourcen: </b> Prozesse verfügen über eigene Ressourcen, wie zum Beispiel Dateien, Netzwerkverbindungen, ect.
- <b>Eigener Kontrollfluss:</b> Jeder Prozess hat einen eigenen Kontrollfluss
- <b>Programmzähler:</b> Das Betriebssystem merkt sich für jeden Prozess die aktuelle Stelle im Programm, die üblicherweise als Adresse im Register "Programm Counter" verzeichnet ist


<h3 style="color:#32CD32">Scheduling und CPU-Ressourcenverteilung</h3>

<u>Grundkonzept Scheduling:</u> Beim Scheduling teilt ein spezieller Systemprozess, bekannt als Dispatcher(auch Scheduler), die CPU-Ressourcen den verschiedenen Prozessen und Threads zu.

<u>Der Prozess kann hierbei folgende <b>Prozesszustände</b> einnehmen, die er durchläuft:</u>
- <b>Ready:</b> Der Prozess ist bereit zur Ausführung, wartet aber auf die Zuteilung der CPU
- <b>Running:</b> Der Prozess wird aktuell von der CPU ausgeführt
- <b>Blocked:</b> Der Prozess wartet auf das Eintreten eines ereignisses oder das Ende eines blockierenden Systemaufrufs und kann in dieser Zeit nicht ausgeführt werden.

<u>Wechsel der Zustände:</u> Prozesse wechseln zwischen diesen Zuständen basierend auf verschiedenen Ereignissen, wie der Fertigstellung eines Systemaufrufs oder dem Ablaufen der ihnen zugewiesenen Zeitscheibe.<br>
<u>Abhängigkeit vom Betriebssystem:</u> Das genaue Zustandsdiagramm und das Verhalten der Prozesse beim Schduling können je nach Betriebssystem variieren. Einige Systeme, wie die JVM,  haben eigene Zustandsdiagramme und Scheduling-Methoden.

Das Scheduling ist abschließend einer der zentralen Bestandteile der betriebssystemfunktionalität, der sicherstellt, dass die Prozessorressourcen effizient und fair zwischen allen Prozessen verteilt werden, um eine optimale Leistung des systems zu gewährleisten.

<u>Erweiternd lassen sich bestimmte <b>Scheduling-Verfahren</b> anführen:</u>
- <b>Präemptives Multitasking mit Round Robin (PRR):</b> Dies ist ein Scheduling-Verfahren, bei dem jedem Prozess eine feste Zeitscheibe (Quantum) zugewiesen wird. Nach Ablauf der Zeitscheibe wird der aktuelle Prozess untebrochen und der nächste bereite Prozess erhält CPU-Zeit.
- <b>Präemptives Multitasking mit Prioritäten (PP):</b> Hierbei werden Prozesse basierend auf ihren Prioritäten geplant. Prozesse mit höherer Priorität(niedrigere prioritätsnummer) erhalten vorrang vor Prozessen mit niedrigerer Priorität

```python
import matplotlib.pyplot as plt

# Entnommen aus Daten
processes = {
    'C': {'Startzeit': 50, 'Rechenzeit': 120},
    'B': {'Startzeit': 17, 'Rechenzeit': 317},
    'A': {'Startzeit': 0, 'Rechenzeit': 60}          
}

# Quantum festlegen
quantum = 20

# Farben für die Prozesse definieren
colors = {'A': 'red', 'B': 'green', 'C': 'blue'}

# Höhe der Balken
bar_height = 0.4

# Berechnen der Round Robin Zeitscheiben
def calculate_rr_schedule(processes, quantum):
    remaining_times = {p: times['Rechenzeit'] for p, times in processes.items()}
    time = 0
    schedule = []
    time_points = set()

    while any(remaining_times.values()):
        for process, times in processes.items():
            if remaining_times[process] > 0 and times['Startzeit'] <= time:
                duration = min(quantum, remaining_times[process])
                end_time = time + duration
                schedule.append((process, time, duration))
                time_points.update([time, end_time])
                remaining_times[process] -= duration
                time += duration
    return schedule, time_points

# Berechnen des Schedules und der Zeitpunkte
schedule, time_points = calculate_rr_schedule(processes, quantum)

# Erstellen des Plots
fig, ax = plt.subplots(figsize=(10, 5))

# Hinzufügen der Zeitscheiben als horizontale Balken
for process, start, duration in schedule:
    color = colors.get(process, 'skyblue')  # Standardfarbe falls nicht definiert
    ax.barh([process], [duration], left=[start], height=bar_height, color=color, align='center')

# Gestrichelte Linien für Start- und Endzeiten
for time in time_points:
    ax.axvline(x=time, color='gray', linestyle='dashed', linewidth=0.5)

# Einstellen der x-Achsenbeschriftungen und -ticks
ax.set_xticks(list(time_points))  # Ticks bei Start- und Endzeiten
ax.set_xticklabels([f'{x}' for x in time_points])  # Beschriftungen der Ticks

# Titel und Beschriftungen
plt.title('Zeitstrahl nach PRR')
plt.xlabel('Zeit (ms)')
plt.ylabel('Prozesse')

# Anzeigen des Plots
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Entnommen aus Daten
processes = {
    'A': {'Startzeit': 0, 'Rechenzeit': 60, 'Priorität': 3},
    'B': {'Startzeit': 17, 'Rechenzeit': 317, 'Priorität': 2},
    'C': {'Startzeit': 50, 'Rechenzeit': 120, 'Priorität': 1}          
}

# Farben für die Prozesse definieren
colors = {'A': 'red', 'B': 'green', 'C': 'blue'}

# Höhe der Balken
bar_height = 0.4

# Berechnen der PP Zeitscheiben
def calculate_pp_schedule(processes):
    time = 0
    schedule = []
    time_points = set()
    process_queue = sorted(processes.items(), key=lambda x: (x[1]['Priorität'], x[1]['Startzeit']))

    while process_queue:
        for i, (process, times) in enumerate(process_queue):
            if times['Startzeit'] <= time:
                schedule.append((process, time, times['Rechenzeit']))
                time_points.update([time, time + times['Rechenzeit']])
                time += times['Rechenzeit']
                process_queue.pop(i)
                break
        else:
            time += 1  # Zeit erhöhen, wenn kein Prozess gestartet werden kann
    return schedule, time_points

# Berechnen des Schedules und der Zeitpunkte
schedule, time_points = calculate_pp_schedule(processes)

# Erstellen des Plots
fig, ax = plt.subplots(figsize=(10, 5))

# Hinzufügen der Zeitscheiben als horizontale Balken
for process, start, duration in schedule:
    color = colors.get(process, 'skyblue')  # Standardfarbe falls nicht definiert
    ax.barh([process], [duration], left=[start], height=bar_height, color=color, align='center')

# Gestrichelte Linien für Start- und Endzeiten
for time in time_points:
    ax.axvline(x=time, color='gray', linestyle='dashed', linewidth=0.5)

# Einstellen der x-Achsenbeschriftungen und -ticks
ax.set_xticks(list(time_points))  # Ticks bei Start- und Endzeiten
ax.set_xticklabels([f'{x} ms' for x in time_points])  # Beschriftungen der Ticks

# Titel und Beschriftungen
plt.title('Zeitstrahl nach PP')
plt.xlabel('Zeit (ms)')
plt.ylabel('Prozesse')

# Anzeigen des Plots
plt.tight_layout()
plt.show()
```
