## Unternehmen und Umwelt

### Typologie des Unternehmens

<h4 style="color:#8FBC8F">An welchen Eigenschaften lässt sich ein Unternehmen typologisch einschätzen?</h4>

Zu den typologischen Eigenschaften eines Unternehmens zählen:

- Gewinnorientierung(Profit vs. Non-Profit)
- Branche/Branchengliederung (Dienstleistung oder Sachleistung)
- Unternehmensgröße
- Technisch-ökonomische Strukturen
- Rechtsform
- Unternehmensverbindungen
- Lebenszyklus
- Standort und Internationalisierung

<h4 style="color:#8FBC8F">Branchengliederung</h4>

Durch die Branchengliederung lassen sich Unternehmen im Wesentlichen in 2 Branchen unterscheiden: 

<u>Dienstleistungsunternehmen: </u> Dienstleistungsunternehmen bieten eine Dienstleistung an. Sie können andere Unternehmen supporten oder aber auch direkt die privaten Haushalte.

<u>Sachleistungsunternehmen: </u> Produzieren durch Erschließung natürlicher Ressourcen und verarbeiten diese sukzessive über mehrere Produktionsschritte zu einem Endprodukt.

<h4 style="color:#8FBC8F">Technisch-ökonomische Struktur</h4>

Mit der technisch-ökonomischen Struktur eines Unternehmens lässt sich ganz grob die tatsächliche Funktionsweise des Unternehmens umreißen. Dazu betrachtet man: 
- vorherschender Produktionsfaktor
- Fertigungstyp
- Anordung der Maschinen

<u>Produktionsfaktoren:</u> Produktionsfaktoren entsprechen immer einer Intensivität. Sie sind zu kategorisieren in:
- personalintensiv
- anlagenintensiv
- materialintensiv
- energieintensiv
- wissensintensiv

<u>Fertigungstypen:</u> Es lässt sich unterscheiden, ob in Einzelfertigung oder in Mehrfachfertigung produziert wird.
<u>Anordnung der Maschinen:</u> Die Anordnung der Maschinen erfolgt nach zwei Prinzipien:
- Werkstattprinzip: (Zeug wird hin und her transportiert)
- Fließbandprinzip: (Zeug wird optimiert durchverarbeitet, ohne lange Strecken zurücklegen zu müssen)

<h4 style="color:#8FBC8F">Rechtsformen</h4>

Ein Unternehmen lässt sich immer einer bestimmten Rechtsform zuweisen. Unterschiedliche Rechtsformen unterscheiden sich hierbei im Wesentlichen durch steuerliche und juristische Aspekte.

<h5 style="color:#708090">Einzelunternehmen</h5>
Bei einem Einzelunternehmen ist eine natürliche Person der einzige MItarbeiter und Inhaber. Der Inhaber haftet hierbei auch mit seinem Privatvermögen.
<h5 style="color:#708090">Gesellschaften</h5>

- Das zur Gründung benötigte Kapital muss konstant bereitgehalten werden

<h5 style="color:#708090">Aufbau einer AG</h5>

Im Nachfolgenden geht es um den Aufbau einer AG. Diese Hat folgende Organe:

- Vorstand: leitet Unternehmen
- Aufsichtsrat: geregelt durch Mitbestimmungsgesetz, bestimmt Vorstandsmitglieder, besteht aus Arbeitnehmer- und Kapital-Vertretern (ab 500 Mitarbeiter: 33% Arbeitnehmervertretung, ab 2000 Mitarbeitern: 50% Arbeitnehmervertretung)
- Hauptversammlung: grundlegende und Unternehmensweite Entscheidungen werden getroffen, sie setzt sich aus Aktionären zusammen. Es wird der Kapitalvertreter der AG gewählt

Die Personengesellschaften haben meist folgende Eigenschaften:

- können von natürlichen oder juristischen Personen gegründet werden.
- ist selbst keine juristische Person
- die Inhaber haften
- solange kein Großunternehmen: keine Publizitätspflicht/Prüfungspflicht
- keine gesetzliche Mitbestimmung für Arbeitnehmer

<h5 style="color:#708090">Gewinnverteilung der OHG</h5>

Die Gewinnverteilung in einer OHG sieht wie folgt aus:

- alle Gesellschafter kriegen 4% Zinsen auf ihr Eigenkapital aus dem Gewinn
- vom Restgewinn erhalten alle Gesellschafter den gleichen Anteil
- Bei Kommanditgesellschaften gilt selbiges, nur die 4% sind dort eher waage reglementiert. Deshalb sollte dieses Verhältnis im Gesellschaftsvertrag geregelt werden


<h4 style="color:#8FBC8F">Unternehmensverbindungen</h4>
Es gibt folgende Arten der Verbindung zu anderen Unternehmen:

- Konsortium: zweckgebundene Vereinigung mehrerer selbstständiger Unternehmen in Form einer GbR
- Kartell: Verbindung zur Begrenzung des Wettbewerbs. Die wirtschaftliche Selbsständigkeit ist teilweie eingeschränkt.(Form für Unternehmen mit Marktmacht meist illegal)

<h4 style="color:#8FBC8F">Standort und Internationalisierung</h4>

Der Standort eines Unternehmens lässt sich anhand des Ausbreitungsgrades kategorisieren. Es gibt hierbei folgende Grade:

1. Lokal
2. Regional
3. National
4. International
5. Multinational

Im Wesentlichen unterscheiden sich die Grade "national" und "multinational" darin, dass bei "international" nur die Produktion im Ausland stattfindet, bei "multinational" wiederum auch der Vertrieb.
Der Internationalisierungsgrad kategoriesiert hier nochmal genauer:
1. Nationale Einbetrieb-Unternehmen
2. Nationale Mehrbetrieb-Unternehmen
3. Internationale Merhbetrieb-Unternehmen (Kein Vertrieb im Ausland)
4. Internationale Mehrbetrieb-Unternehmen (Vertrieb auch im Ausland)

Um eine Internationalisierung zu erreichen gibt es verschiedene "Internationalisierungsstrategien". Einige Beispiele:

- Export ins Ausland
- Produktion im Ausland
- Lizensverträge, sodass ausländische Unternehmen das Selbe Produkt gegen Lizenzgebühren anbieten dürfen
- Franchising zur Dienstleistungsanbietung im Ausland
- Auslandsniederlassung (rechtich unselbstständig)
- Tochtergesellschaft
